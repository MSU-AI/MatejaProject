```python
import pandas as pd
```

```python
#FIRST YOU LOAD IN THE FILE USING PANDAS
df = pd.read_csv('Monopoly/Data/board.csv', delimiter = ',')
df
```

```python
#WE WANT TO CREATE TWO LISTS THAT HOLD X AND Y POSITIONS AS WE NEED TO ADJUST THE CSV FILE FOR OUR PROJECT
#AS THE BOARD IS NOT LINEAR BUT 2D(4 ARRAYS, ONE FOR EACH SIDE)
pos_x = []
pos_y = []
for s in range(df.shape[0]):
    pos_y.append(df.iloc[s]['Position']%10)
    pos_x.append(int(df.iloc[s]['Position'] / 10))
    
```

```python
#THEN WE JUST UPDATE COLUMNS AND INSERT A NEW ONE
df['Position'] = pos_y
df = df.rename(columns={"Position" : "Position(Y)"})
df.insert(3, 'Position(X)', pos_x)
df
```
